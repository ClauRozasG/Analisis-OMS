# **Machine Learning**

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
x = df.drop(columns = ['Life expectancy ', 'Country'])
x
```

```python
y=df[['Life expectancy ']]
y
```

```python
y.shape
```

```python
x.shape
```

```python
#Convertimos los valores a float
x=np.array(x).astype('float32')
y=np.array(y).astype('float32')
```

```python
seed=42
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.25)
model=LinearRegression()
model.fit(x_train,y_train)
previsiones=model.predict(x_test)
tasa_acierto=mean_squared_error(y_test, previsiones)

print(f'El error cuadratico medio fue de: {round(tasa_acierto,2)}')
```

```python
import xgboost as xgb
model = xgb.XGBRegressor(objective ='reg:squarederror', learning_rate = 0.1, max_depth = 30, n_estimators = 100)
model.fit(x_train, y_train)
```

```python
model_accuracy = model.score(x_test, y_test)
model_accuracy
```

```python
y_pred = model.predict(x_test)
y_pred
```

```python
plt.plot(y_test, y_pred, "^", color = 'orange')
plt.xlabel('Valores verdaderos')
plt.ylabel('Predicciones')
plt.show()
```
